```python
# Step 1 Load the data
import pandas as pd
```

```python
nflx_data = pd.read_csv('/content/sample_data/NFLX.csv',index_col='Date')
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import datetime as dt

d=''
plt.figure(figsize= (15,10))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=60))
x_dates = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in nflx_data.index.values]

plt.plot(x_dates, nflx_data['High'], label='High')
plt.plot(x_dates, nflx_data['Low'], label='Low')
plt.xlabel('Time Scale')
plt.ylabel('Scaled USD')
plt.legend()
plt.gcf().autofmt_xdate()
plt.show()
```
